```python
df = data.copy()
```

```python
df= df.rename(columns = {"ID":"IdenNum"})
```

```python
df.describe()
```

```python
#Drop "Income" outlier.
df = df.dropna(subset=["Income"])
df = df[df["Income"]<600000]
```

```python
df.describe()
```

```python
s = (df.dtypes == "object")
object_cols = list(s[s].index)

print("Categorical variables in the dataset:", object_cols)
```

```python
df["Average"].fillna(df["Average"].mean(), inplace = True)
```

```python
df["Average"] = np.where(df["Average"]> 200,200, df["Average"])
```

```python
df.drop("IdenNum", axis=1)
```

```python
df["Children"].replace({"3 children": 3, "2 children" : 2 ,"1 child": 1, "No child" : 0},inplace=True)
```

```python
df.drop("ChildCount", axis=1)
```

```python
clust_cus = df[["Average","CustomerStatus","TotalPurchases"]].copy()
scaler = StandardScaler()

for i in clust_cus.columns:
    
    clust_cus[i] = scaler.fit_transform(np.array(clust_cus[[i]]))
```

```python
cost = []
for i in range(1,20,2) :
    kclust = KMeans(n_clusters = i, random_state = 13)
    kclust.fit(clust_cus)
    cost.append(kclust.inertia_)
```

```python
plt.figure(figsize = (12,8))
plt.title("The Elbow", size =25, y = 1.03, fontname = "monospace")
plt.grid(color = "grey", linestyle = ":", axis = "y", alpha =0.8, zorder = 0, dashes = (1,7))
a = sns.lineplot(x=range(1,20,2),y = cost, linewidth = 3)
sns.scatterplot(x = range(1,20,2), y = cost, s = 60, edgecolor = "black", zorder = 5)
plt.xlabel("Number of clusters", size =14, fontname = "monospace")
plt.ylabel("COST", size =14, fontname = "monospace")


plt.annotate('''Optimal number
of clusters''', xy = (5.0, 2400), xytext = (5, 2800),
             arrowprops = dict(facecolor = 'steelblue', arrowstyle = "->", connectionstyle = "arc3,rad = 0.4",color = '#dd4124'), 
             fontsize = 13, fontfamily = 'monospace', ha = 'center', color = '#dd4124')
plt.show()
```

```python
gmm = GaussianMixture(n_components=4, covariance_type='spherical',max_iter=3000, random_state=13).fit(clust_cus)
labels = gmm.predict(clust_cus)
clust_cus["Cluster"] = labels
re_clust = {
        0: "Elite Customer",
        1: "Economical Customer",
        2: "Potential Customer",
        3: "Good Customer"
}

clust_cus["Cluster"] = clust_cus["Cluster"].map(re_clust)
```

```python
df = df.merge(clust_cus.Cluster, left_index=True, right_index=True)
```

```python
pd.options.display.float_format = "{:.0f}".format
summary = df[["Income","Spending","CustomerStatus","Cluster"]]
summary.set_index("Cluster", inplace = True)
summary = summary.groupby("Cluster").describe().transpose()
summary
```

```python
clusters_products = df[['Wines', 'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold', 'Cluster']]    # Select variables

clusters_products1 = clusters_products.groupby(['Cluster'])
clusters_products2 = clusters_products1.agg({'Wines':'sum', 'Fruits':'sum', 'Meat':'sum', 'Fish':'sum', 'Sweets':'sum', 'Gold':'sum'})

clusters_products3 = clusters_products2.stack().reset_index(name='Count').rename(columns={'level_1':'Products'})   # Oposite as pivoting

clusters_products3['group'] = clusters_products3['Cluster']
clusters_products3['group'] = clusters_products3['group'].astype(str)

# Rename values
clusters_products3['group'] = clusters_products3['group'].str.replace('0', 'Good Customers')
clusters_products3['group'] = clusters_products3['group'].str.replace('1', 'Elite Customers')
clusters_products3['group'] = clusters_products3['group'].str.replace('2', 'Potential Customers')
clusters_products3['group'] = clusters_products3['group'].str.replace('3', 'Economical Customers')

products = clusters_products3.copy()
products = products.assign(ratio=products.groupby('group').Count.transform(lambda x: x / x.sum()))

# Visualization
fig = px.bar(products, x='group', y='ratio', color='Products',
             labels={
                     "ratio": "Ratio",
                     "group": "Consumer's type"
                     },
             color_discrete_map={
                     'Gold': '#FFD700',
                     'Fish': '#87CEEB',
                     'Wines': '#b11226',
                     'Meat': '#f08080',
                     'Sweet': '#FF69B4',
                     'Fruits': 'lightgreen'},
                title="Products Distribution by Clusters")

fig.layout.yaxis.tickformat = ',.0%'

fig.update_traces(marker_line_color='white', marker_line_width=1, opacity=0.8)

fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)

fig.update_layout(
    {'plot_bgcolor': 'black',
    'paper_bgcolor': 'black'
    },
    font=dict(
        family="verdana",
        size=21,
        color="white"
    ),
    width=680,
    height=800,
    title_font_color="#FFC300",
    yaxis_title=None,
    xaxis_title=None
)

fig.show()
```
