```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
raw = "Resources/raw.csv"

raw_data = pd.read_csv(raw, sep='\t')
```

```python
raw_data["Marital_Status"].value_counts()
```

```python
raw_data["Marital_Status"] = raw_data["Marital_Status"].replace("Alone", "Single")
raw_data["Marital_Status"] = raw_data["Marital_Status"].replace("Absurd", "Single")
raw_data["Marital_Status"] = raw_data["Marital_Status"].replace("YOLO", "Single")
```

```python
raw_data["Education"].value_counts()
```

```python
raw_data.columns
```

```python
raw_data["Age"] = 2021-raw_data["Year_Birth"]
```

```python
raw_data['Dt_Customer'] =  pd.to_datetime(raw_data['Dt_Customer'])
```

```python
raw_data_free = raw_data.dropna()
```

```python
data = raw_data_free.copy()
```

```python
data.count()
```

```python
correlation = data.corr()
plt.figure(figsize = (25,25))
sns.heatmap(correlation, annot = True)
plt.show()
```

```python
unstack_corr = correlation.unstack()
unstack_corr.sort_values(ascending = False)
pd.set_option('display.max_columns',1000)
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_colwidth', -1)
unstack_corr
```

```python
data["Expenses"] = data["MntWines"]+ data["MntFruits"]+ data["MntMeatProducts"]+ data["MntFishProducts"]+ data["MntSweetProducts"]+ data["MntGoldProds"]
```

```python
data["Total_Purchases"] = data["NumDealsPurchases"]+data["NumWebPurchases"]+data["NumCatalogPurchases"]+data["NumStorePurchases"]
```
