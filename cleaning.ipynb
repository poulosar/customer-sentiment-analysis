```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
raw = "Resources/raw.csv"

raw_data = pd.read_csv(raw, sep='\t')
```

```python
raw_data["Marital_Status"] = raw_data["Marital_Status"].replace("Alone", "Single")
raw_data["Marital_Status"] = raw_data["Marital_Status"].replace("Absurd", "Single")
raw_data["Marital_Status"] = raw_data["Marital_Status"].replace("YOLO", "Single")
```

```python
raw_data["Age"] = 2021-raw_data["Year_Birth"]
```

```python
raw_data['Dt_Customer'] =  pd.to_datetime(raw_data['Dt_Customer'])
```
